```python
import json
from pathlib import Path
from datasets import Dataset, DatasetDict

# Get the size of the dataset
print("Dataset structure:")
print(abstention_bench_data)
print(f"\nDataset type: {type(abstention_bench_data).__name__}")

# Handle both Dataset and DatasetDict
if isinstance(abstention_bench_data, DatasetDict):
    print("\nDataset splits:", list(abstention_bench_data.keys()))
    total_size = 0
    for split_name, split_data in abstention_bench_data.items():
        split_size = len(split_data)
        total_size += split_size
        print(f"{split_name}: {split_size:,} examples")
    print(f"\nTotal size: {total_size:,} examples")
    
    # Save as .jsonl file in data/ directory
    output_path = Path("data/abstention_bench_data.jsonl")
    output_path.parent.mkdir(parents=True, exist_ok=True)
    
    with open(output_path, 'w', encoding='utf-8') as f:
        for split_name, split_data in abstention_bench_data.items():
            for example in split_data:
                # Convert to dict and add split information
                example_dict = dict(example)
                example_dict['split'] = split_name
                f.write(json.dumps(example_dict, ensure_ascii=False) + '\n')
    
    print(f"\nSaved {total_size:,} examples to {output_path}")
    
elif isinstance(abstention_bench_data, Dataset):
    total_size = len(abstention_bench_data)
    print(f"\nDataset size: {total_size:,} examples")
    
    # Save as .jsonl file in data/ directory
    output_path = Path("data/abstention_bench_data.jsonl")
    output_path.parent.mkdir(parents=True, exist_ok=True)
    
    with open(output_path, 'w', encoding='utf-8') as f:
        for example in abstention_bench_data:
            # Convert to dict
            example_dict = dict(example)
            f.write(json.dumps(example_dict, ensure_ascii=False) + '\n')
    
    print(f"\nSaved {total_size:,} examples to {output_path}")
else:
    print(f"Unexpected dataset type: {type(abstention_bench_data)}")
```

```python
# Read the .arrow file
import pyarrow.parquet as pq
import pyarrow as pa
from pathlib import Path

arrow_file_path = Path("data/falseqa/data-00000-of-00001.arrow")

# Method 1: Using datasets library (recommended for HuggingFace datasets)
try:
    falseqa_data = datasets.load_from_disk(str(arrow_file_path.parent))
    print(f"Loaded dataset using datasets library")
    print(f"Dataset type: {type(falseqa_data).__name__}")
    
    # Handle both Dataset and DatasetDict
    if isinstance(falseqa_data, datasets.DatasetDict):
        print(f"\nDataset splits: {list(falseqa_data.keys())}")
        for split_name, split_data in falseqa_data.items():
            print(f"{split_name}: {len(split_data):,} examples")
        # Use the first split for examples
        data_for_examples = list(falseqa_data.values())[0]
    else:
        data_for_examples = falseqa_data
    
    num_questions = len(data_for_examples)
    print(f"\nTotal number of questions: {num_questions:,}")
    
    # Show dataset features/columns
    print(f"\nDataset features/columns: {data_for_examples.column_names}")
    
    # Show some example questions
    print(f"\n{'='*80}")
    print("Example questions (first 3):")
    print(f"{'='*80}")
    
    for i in range(min(3, num_questions)):
        example = data_for_examples[i]
        print(f"\n--- Example {i+1} ---")
        for key, value in example.items():
            if isinstance(value, str) and len(value) > 200:
                print(f"{key}: {value[:200]}...")
            else:
                print(f"{key}: {value}")
        print()
        
except Exception as e:
    print(f"Error loading with datasets library: {e}")
    print("\nTrying with pyarrow directly...")
    
    # Method 2: Using pyarrow directly
    try:
        with pa.memory_map(str(arrow_file_path), 'r') as source:
            table = pa.ipc.open_file(source).read_all()
        
        num_questions = len(table)
        print(f"\nTotal number of questions: {num_questions:,}")
        print(f"Columns: {table.column_names}")
        
        # Convert to pandas for easier viewing
        df = table.to_pandas()
        print(f"\n{'='*80}")
        print("Example questions (first 3):")
        print(f"{'='*80}")
        
        for i in range(min(3, num_questions)):
            print(f"\n--- Example {i+1} ---")
            row = df.iloc[i]
            for col in df.columns:
                value = row[col]
                if isinstance(value, str) and len(value) > 200:
                    print(f"{col}: {value[:200]}...")
                else:
                    print(f"{col}: {value}")
            print()
            
    except Exception as e2:
        print(f"Error loading with pyarrow: {e2}")
```
